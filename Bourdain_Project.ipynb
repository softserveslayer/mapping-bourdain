```python
import pandas as pd

data = pd.read_csv('bourdain_travel_places.csv')
datalist = data.values.tolist()
```

```python
short_datalist = datalist[0:50]
clean_datalist = [data for data in datalist if not pd.isnull(data[7]) if not pd.isnull(data[4])]

# if not pd.isnull(atext):
#   for word in nltk.word_tokenize(atext):
# clean_datalist
```

```python
locations = []
for location in clean_datalist:
    coords = location[7].split(",")
    noapostraphecity = location[4].replace("'","")
    
    locations.append({'season' : location[1], 'episode' : location[2], 'title' : location[3], 
                       'text': f"show: {location[0]}, title: {noapostraphecity}, season {location[1]}, episode {location[2]}", 
                      'country' : location[5], 'state' : location[6], 
                      'coordinates' : list(map(float,coords))})

#     locations.append({'season' : location[1], 'episode' : location[2], 'title' : location[3], 
#                       'text': location[0] + '\n' +
#                        location[4].replace("'","") + '\n' + "Season" + str(location[1]) + '\n' + "Episode" + str(location[2]), 
#                       'country' : location[5], 'state' : location[6], 
#                       'coordinates' : list(map(float,coords))})
locations
```
